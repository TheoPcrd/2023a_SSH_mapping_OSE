```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,30,10000)
k = 1.5
l = 3
f = np.sin(2*k*np.pi*x) + np.sin(2*l*np.pi*x) + np.random.randn(x.shape[0])*0.001
```

```python
plt.plot(x,f)
```

```python
x[1]-x[0]
```

```python
k = 2*np.pi*np.fft.fftfreq(x.shape[0],x[1]-x[0])
fft = np.fft.fftshift(np.fft.fft(f))
plt.plot(k,fft)
plt.xlim([-10,10])
```
